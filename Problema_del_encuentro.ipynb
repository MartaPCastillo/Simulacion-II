```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def hola(n):
  X=[]
  Y=[]
  favorable = 0
  i = 0
  Lista = []
  for i in range (n):
    X.append (random.uniform(0,30))
    Y.append(random.uniform(0,30))
    if abs(X[i]-Y[i]) <= 10:
        favorable +=1
        i += 1
        probabilidad = favorable/(i)
        Lista.append(probabilidad)
  return Lista
```

```python
print(X)
print(Y)
```

```python
Nf = hola(900)
print(Nf)
```

```python
np.mean(Nf)
```

```python
plt.plot(Nf)
plt.xlabel('p')
plt.ylabel('iteraciones')
plt.title('Prob. de que el tiempo que espera una persona a otra sea max. 10')
plt.show()
```
